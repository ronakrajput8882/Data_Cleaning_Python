```python
import json
```

```python
with open("RAW_DATA.txt", "r") as f:
    data = json.load(f)
```

```python
bad_values = ["ERROR", "UNKNOWN", None]

for row in data:
    for key in row:
        if row[key] in bad_values:
            row[key] = None
```

```python
def to_int(x):
    try:
        return int(x)
    except:
        return None

def to_float(x):
    try:
        return float(x)
    except:
        return None

for row in data:
    row["Quantity"] = to_int(row["Quantity"])
    row["Price Per Unit"] = to_float(row["Price Per Unit"])
    row["Total Spent"] = to_float(row["Total Spent"])
```

```python
for row in data:
    if row["Quantity"] is not None and row["Price Per Unit"] is not None:
        row["Total Spent"] = row["Quantity"] * row["Price Per Unit"]
```

```python
clean_data = []

for row in data:
    if (
        row["Quantity"] is not None and
        row["Price Per Unit"] is not None and
        row["Transaction Date"] is not None
    ):
        clean_data.append(row)

data = clean_data
```

```python
with open("CLEAN_DATA.txt", "w") as f:
    json.dump(data, f, indent=2)
```
